<br><br>MULTIPLE LINEPLOT

```python
day = [1,2,3,4]

vivo_sales = [25000,45999,23900,12000]
samsung_sales = [45000,89345,35429,14977]
apple_sales = [120000,7000,290455,30000]
oppo_sales = [12000,70000,2300,90080]

#to set the figure size:
plt.figure(figsize = (15, 8))

#to execute the plot:
plt.plot(day, vivo_sales, marker ='*', color = 'g', label = 'vivo', ls = ':', lw = 3)
plt.plot(day, samsung_sales, marker ='h', color = 'b', label = 'samsung', ls = ':', lw = 3)
plt.plot(day, apple_sales, marker ='x', color = 'y', label = 'apple', ls = ':', lw = 3)
plt.plot(day, oppo_sales, marker ='D', color = 'r', label = 'oppo', ls = ':', lw = 3)

#to add titles:
plt.xlabel(f'Days', fontweight = 'bold', fontsize = '12')
plt.ylabel(f'Sales', fontweight = 'bold', fontsize = '12')
plt.title(f'Sales Over different Days', fontweight = 'bold', fontsize = '12')

plt.legend(title = f'Products', loc = 'upper right', fontsize = 14)

plt.xticks([1,2,3,4])

plt.show()
```

Bar Charts

1. Vertical Bar Chart
2. Horizontal Bar Chart
3. Stacked Bar Chart

```python
month = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
sales = [12300,13289,15634,14588,6587,5329,37400,25356,54783,34890,54278,43648]

plt.figure(figsize=(15,5))
bars = plt.bar(month,sales, color = 'pink', edgecolor = 'purple', linewidth = 3)

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height, #y
        f'{height}',
        ha = 'center', va = 'bottom', fontweight = 'bold', fontsize = 10
    )

plt.xlabel(f'Months', fontweight = 'bold', fontsize = '12')
plt.ylabel(f'Sales', fontweight = 'bold', fontsize = '12')
plt.title(f'Sales Trend Over Months', fontweight = 'bold', fontsize = '12')

plt.show()
```

<br><br>Horizontal Bar Chart

```python
month = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
sales = [12300,13289,15634,14588,6587,5329,37400,25356,54783,34890,54278,43648]

plt.figure(figsize=(12,7))
bars = plt.barh(month,sales, color = 'pink', edgecolor = 'purple', linewidth = 3)

for i, value in enumerate (sales):
    plt.text(value +500, i,
             f'{str(value)}',
             va = 'center_baseline',
             fontweight = 'bold', fontsize = 12
            )

plt.xlabel(f'Months', fontweight = 'bold', fontsize = '12')
plt.ylabel(f'Sales', fontweight = 'bold', fontsize = '12')
plt.title(f'Sales Trend Over Months', fontweight = 'bold', fontsize = '12')

plt.show()
```

Grouped Bar Plot

```python
categories = ['Q1','Q2','Q3','Q4']
product_A = [450,390,1240,690]
product_B = [360,215,943,345]

x = np.arange(len(categories))
width = 0.35

plt.figure(figsize=(15,6))
bar1 = plt.bar(x - width / 2,product_A,width, label = 'Product A', color = 'pink')
bar2 = plt.bar(x + width / 2,product_B,width, label = 'Product B', color = 'purple')

for bar in bar1 + bar2:
    plt.text(
        bar.get_x() + bar.get_width() / 2, #x
        bar.get_height() + 10, #y
        f'${bar.get_height() + 4}', #s 
        ha = 'center', va = 'bottom', fontweight = 'bold', fontsize = 12
    )

plt.xticks(ticks = x,
           labels = ['Quarter - 1','Quarter - 2','Quarter - 3','Quarter - 4'],
           fontweight = 'bold', fontsize = 10)

plt.yticks([])

plt.xlabel(f'Periodic Analysis', fontweight = 'bold', fontsize = '12')
plt.ylabel(f'Sales', fontweight = 'bold', fontsize = '12')
plt.title(f'Sales Trend - Quarter', fontweight = 'bold', fontsize = '12')

plt.legend(title = 'PRODUCTS', edgecolor = 'black', fontsize = 12, shadow = True)

plt.show()
```

```python
categories = ['Q1','Q2','Q3','Q4']
product_A = [540,390,1240,690]
product_B = [360,215,943,345]

x = np.arange(len(categories))
width = 0.35

plt.figure(figsize=(15,6))
bar1 = plt.bar(x,product_A, width = 0.6, label = 'Product A', color = 'pink', edgecolor = 'black', lw = 3)
bar2 = plt.bar(x,product_B, width = 0.6, bottom = product_A, label = 'Product B', color = 'purple', edgecolor = 'black', lw = 3)

for i, value in enumerate(product_A):
    plt.text(i,  #x
             value /2,  #y
             f'${str(value)} - Prod A',    #s
             ha = 'center', va = 'bottom',
             fontsize = 12, fontweight = 'bold', color = 'black')
    
for i, value in enumerate(product_B):
    plt.text(i,  #x 
             product_A[i] + value /2,  #y
             f'${str(value)} - Prod B',
             ha = 'center', va = 'bottom',
             fontsize = 12, fontweight = 'bold', color = 'black')

plt.xticks(ticks = x,
           labels = ['Quarter - 1','Quarter - 2','Quarter - 3','Quarter - 4'],
           fontweight = 'bold', fontsize = 10)

plt.yticks([])

plt.xlabel(f'Periodic Analysis', fontweight = 'bold', fontsize = '12')
plt.ylabel(f'Sales in (USD)', fontweight = 'bold', fontsize = '12')
plt.title(f'Sales Trend - Quarter', fontweight = 'bold', fontsize = '12')

plt.legend(loc = 'upper left', title = 'PRODUCTS', edgecolor = 'black', fontsize = 12, shadow = True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

categories = ['Q1','Q2','Q3','Q4']
product_A = [540,390,1240,690]
product_B = [360,215,943,345]
product_C = [379,415,824,654]

x = np.arange(len(categories))

plt.figure(figsize=(15,6))

# Bars
bar1 = plt.bar(x, product_A, width=0.6, label='Product A',
               color='pink', edgecolor='black', lw=3)

bar2 = plt.bar(x, product_B, width=0.6, bottom=product_A, label='Product B',
               color='purple', edgecolor='black', lw=3)

bar3 = plt.bar(x, product_C, width=0.6,
               bottom=np.array(product_A) + np.array(product_B),
               label='Product C',
               color='magenta', edgecolor='black', lw=3)

# Text labels for Product A
for i, value in enumerate(product_A):
    plt.text(i, value/2,
             f'${value} - Prod A',
             ha='center', va='bottom',
             fontsize=12, fontweight='bold', color='black')

# Text labels for Product B
for i, value in enumerate(product_B):
    plt.text(i, product_A[i] + value/2,
             f'${value} - Prod B',
             ha='center', va='bottom',
             fontsize=12, fontweight='bold', color='black')

# Text labels for Product C (corrected)
for i, value in enumerate(product_C):
    plt.text(i, product_A[i] + product_B[i] + value/2,
             f'${value} - Prod C',
             ha='center', va='bottom',
             fontsize=12, fontweight='bold', color='black')

# Axis formatting
plt.xticks(x,
           ['Quarter - 1','Quarter - 2','Quarter - 3','Quarter - 4'],
           fontweight='bold', fontsize=10)

plt.yticks([])

plt.xlabel('Periodic Analysis', fontweight='bold', fontsize=12)
plt.ylabel('Sales in (USD)', fontweight='bold', fontsize=12)
plt.title('Sales Trend - Quarter', fontweight='bold', fontsize=12)

plt.legend(loc='upper left', title='PRODUCTS',
           edgecolor='black', fontsize=12, shadow=True)

plt.show()
```

Pie Chart

```python
countries = ['USA','France','India','Africa','Germany','UK','Russia']
population = [7.8,3.3,5.8,8.3,4.6,2.1,4.4]

plt.figure(figsize = (22,7))
plt.pie(
        x= population, 
        labels = countries, autopct = '%.2f%%',
        colors = ['yellow','green','blue','red','pink','purple','brown'],
        explode = [0.2,0,0,0,0,0,0],
        wedgeprops = {'width' : .6, 'edgecolor' : 'black'},
        shadow = True,
        textprops = {'fontsize' : 7, 'color' : 'black', 'fontweight' : 'bold'}
)

plt.legend(fontsize = 7, frameon = False, title = 'Countries')

plt.title('Country wise Population Distribution', fontsize = 14, fontweight = 'bold')

plt.show()
```

<br><br><br>
Area Chart

```python
months = ['Jan','Feb','Mar','Apr']
sales_A = [350,700,120,450]
sales_B = [150,170,520,80]

plt.figure(figsize=(15,5))
plt.fill_between(months, sales_A, label = 'Sales - A', ls = '--', lw = 2.2, color = 'pink', edgecolor = 'black')
plt.fill_between(months, sales_B, label = 'Sales - B', ls = '--', lw = 1.7, color = 'purple', edgecolor = 'black', alpha = .7)

plt.xlabel('Months', fontsize = 10, fontweight = 'bold')
plt.ylabel('Sales A&B', fontsize = 10, fontweight = 'bold')
plt.title('Sales Over Months', fontsize = 13, fontweight = 'bold')

for i, value in enumerate(months):
    plt.text(value,  #x
             sales_A[i] + 30,   #y
             f'{sales_A[i]}',  #s
             ha = 'center', va = 'bottom', fontweight = 'bold')

for i, value in enumerate(months):
    plt.text(value,  #x
             sales_B[i] + 30,
             f'{sales_B[i]}',
             ha = 'center', va = 'bottom', fontweight = 'bold')
             
plt.legend(fontsize = 12)

plt.yticks([])
plt.xticks(ticks = months, fontweight = 'bold', labels = ['JANUARY','FEBRUARY','MARCH','APRIL'])

plt.show()
```

<br><br><br>Step Plot

```python
months = ['Jan','Feb','Mar','Apr']
sales_A = [350,700,120,450]
sales_B = [150,170,520,80]

plt.figure(figsize=(15,5))
plt.step(months, sales_A, label = 'Sales - A', ls = '--', lw = 2.2, color = 'red', where = 'mid')
plt.step(months, sales_B, label = 'Sales - B', ls = '--', lw = 2.2, color = 'blue', where = 'post')

plt.xlabel('Months', fontsize = 10, fontweight = 'bold')
plt.ylabel('Sales A&B', fontsize = 10, fontweight = 'bold')
plt.title('Sales Over Months', fontsize = 13, fontweight = 'bold')

for i, value in enumerate(months):
    plt.text(value,  #x
             sales_A[i] + 30,   #y
             f'{sales_A[i]}',  #s
             ha = 'center', va = 'bottom', fontweight = 'bold')

for i, value in enumerate(months):
    plt.text(value,  #x
             sales_B[i] + 30,
             f'{sales_B[i]}',
             ha = 'center', va = 'bottom', fontweight = 'bold')
             
plt.legend(fontsize = 12)

plt.yticks([])
plt.xticks(ticks = months, fontweight = 'bold', labels = ['JANUARY','FEBRUARY','MARCH','APRIL'])

plt.show()
```
