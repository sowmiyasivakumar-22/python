**Determine Category-Wise Revenue Growth Year Over Year**

Question – 17 :

- Understanding growth trends across product categories helps in strategic investment decisions.
- The executive team is interested in knowing which product categories are showing positive or negative revenue growth year over year.
- Calculate the total revenue per category for each year, and then derive a feature for Year-over-Year (YoY) Growth (%) to highlight accelerating or declining segments.

```python
revenue = df.groupby(['ProductCategory','year'])['TotalPrice'].sum().reset_index().round(2)
revenue_pivot = revenue.pivot(index = 'ProductCategory',
                              columns = 'year',
                              values = 'TotalPrice')
                              
if(2022 in revenue_pivot.columns and 2023 in revenue_pivot.columns):
    revenue_pivot['YoY growth - (2022 - 2023) %'] = revenue_pivot[2022]-revenue_pivot[2023] / revenue_pivot [2022] * 100
if(2023 in revenue_pivot.columns and 2024 in revenue_pivot.columns):
    revenue_pivot['YoY growth - (2023 - 2024) %'] = revenue_pivot[2023]-revenue_pivot[2024] / revenue_pivot [2023] * 100
revenue_pivot
```

**Identify the Return Rate of High-Value Orders Compared to Low-Value Orders**

Question – 19 :
    
- Customer behavior can vary significantly based on the value of a transaction. 
- The operations and fraud detection teams want to compare the return/cancel rate of high-value orders (orders above the 75th percentile) with that of low-value orders
(orders below the 25th percentile). 
- This can help uncover patterns such as abuse of returns or product dissatisfaction in expensive or cheap items.

```python
q25 = df['TotalPrice'].quantile(0.25)
q75 = df['TotalPrice'].quantile(0.75)

def order_value(x):
    if x <= q25:
        return 'Low - Level'
    elif x>= q75:
        return 'High - Level'
    else:
        return 'Mid - Level'

df['OrderCategory'] = df['TotalPrice'].apply(order_value)

return_rate = df.groupby('OrderCategory').agg(
    TotalOrders = ('OrderID','count'),
    ReturnOrCancelled = ('ReturnOrCancelled','sum')
)
return_rate['Return / Cancel %'] = (return_rate['ReturnOrCancelled'] /
                                    return_rate['TotalOrders']) * 100
return_rate.round(2)
```

**Segment Sellers by Revenue Tiers and Analyze Their Average Completion Rate**

Question – 23 :
    
- To design tiered incentive programs for marketplace sellers, the business team needs to segment sellers into Low, Mid, and High Revenue groups.
- Then, calculate the average order completion rate within each revenue tier. 
- This analysis will reveal whether top sellers are also the most reliable, or if mid/low-tier sellers outperform them in fulfillment quality.

```python
df.IsOrderDelivered
```

```python
df.dtypes
```

```python
seller_stats = df.groupby('SellerID').agg(

    TotalRevenue = ('TotalPrice','sum'),
    TotalOrders = ('OrderID','count'),
    CompletedOrders = ('IsOrderDelivered','sum')
).reset_index()

seller_stats['Completed_rate %'] = (seller_stats['CompletedOrders'] 
                                    / seller_stats['TotalOrders']) * 100
seller_stats

q37 = seller_stats['TotalRevenue'].quantile(0.37)
q67 = seller_stats['TotalRevenue'].quantile(0.67)

def revenue_tier(revenue):
    if revenue <=q37:
        return 'Low'
    elif revenue <=q67:
        return 'Mid'
    else:
        return 'High'
seller_stats['Revenue Tier'] = seller_stats['TotalRevenue'].apply(revenue_tier)
tier_analysis = seller_stats.groupby('Revenue Tier').agg(
    AvgCompletionRate = ('Completed_rate %','mean'),
    SellerCount = ('SellerID','count')
).reset_index()

tier_analysis['Avg Completion Rate (%)'] = (tier_analysis['AvgCompletionRate'] / 
                                            tier_analysis['SellerCount']*100).round(2)
tier_analysis[['Revenue Tier','SellerCount','Avg Completion Rate (%)']]
```
